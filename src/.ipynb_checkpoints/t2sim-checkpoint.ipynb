```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
from epg import *
%matplotlib inline
pylab.rcParams['figure.figsize'] = 10, 4
```

```python
TE = 5.648e-3
angles = np.loadtxt('/Users/jtamir/scan-data/feet/synth_foot/sim_for_paper/flipmod/flipangles.txt.408183520')
T = len(angles)
# angles = 150 * np.ones((T))
angles_rad = np.pi / 180 * angles
```

```python
T1vals = np.array([1000.]) * 1e-3
T2vals = np.array([50., 150., 300., 500.]) * 1e-3
L1 = len(T1vals)
L2 = len(T2vals)

S = np.zeros((T,L1,L2), dtype=complex)
S2 = np.zeros((T,L1,L2), dtype=complex)
for j in range(L1):
    T1 = T1vals[j]
    for k in range(L2):
        T2 = T2vals[k]
        
        P = np.matrix([[0],[0],[1]])    # initially in M0

        P = epg_rf(P, pi/2, pi/2)[0]    # 90 degree tip
            
        s, s2 = FSE_signal2(angles_rad, TE, T1, T2)
        S[:,j,k] = s.ravel()
        S2[:,j,k] = s2.ravel()
        

            
plt.plot(range(T), np.abs(S.reshape((T, -1))), range(T), np.abs(S2.reshape((T, -1))), 'k--')
plt.ylim((0,1))
# plt.legend(['T1 = %d ms' % (1000 * s) for s in T1vals])
```
